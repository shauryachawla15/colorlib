```python
import pandas as pd
import bs4
import requests
import re
```

```python
url=input("Enter Web-URL =   ")
data=requests.get(url)
soup=bs4.BeautifulSoup(data.text,"html.parser")

print("___________________________________________")
print("    WELCOME TO WEB-SCRAPING SYSTEM.        ")
print("___________________________________________")


# FOR PARAGRAPH -------->

def paragraph():
    Para=open("Para.txt","w")
    for para in soup.find_all('p'):
        Para.write("\n")
        Para.write(para.text)
    Para.close()
    print()
    print("Paragraph file created....")
    
    
def headlines():
    while(True):
        ch=input("Enter heading-type. =   ")
        if ch=="h1":
            Head=open("Heading-h1.txt","w")
            for headin soup.find_all("h1"):
                Head.write("\n")
                Head.write(head.text)
            print("Heading-h1 file created.")
            Head.close()
            
        elif ch=="h2":
            Head=open("Heading-h2.txt","w")
            for head in soup.find_all("h2"):
                Head.write("\n")
                Head.write(head.text)
            print("Heading-h2 file created.")
            Head.close()
            
        elif ch=="h3":
            Head=open("Heading-h3.txt","w")
            for headin soup.find_all("h3"):
                Head.write("\n")
                Head.write(head.text)
            print("Heading-h3 file created.")
            Head.close()
            
        elif ch=="h4":
            Head=open("Heading-h4.txt","w")
            for head in soup.find_all("h4"):
                Head.write("\n")
                Head.write(head.text)
            print("Heading-h4 file created.")
            Head.close()
            
        else:
            print("You have Entered wrong choice.....")
            a=input("Do you want to continue this loop(Y/N) ? ")
            if a=="Y":
                continue
                
            else:
                break
                
# FOR WEBLINKS -------->

def weblink():
    Web=open("Weblink.txt","w")
    for link in soup.find_all('a',attrs={"href":re.compile("^https://")}):
        Weblink=link.get("href")
        Web.write("\n")
        Web.write(link.text)
    Web.close()
    print()
    print("Weblink file created....")
            
            
# FOR IMAGES -------->

def image():
    Image=open("imagelink.txt","w")
    for images in soup.find_all('img'):
        images.append(images.get("src"))
        Image.write("\n")
        Image.write(images.text)
    Image.close()
    print()
    print("Imagelink file created....")
    
while(True):
    print("Please choose ny one option.......")
    print("1.Scrape Paragraph")
    print("2.Scrape Headlines")
    print("3.Scrape Weblink")
    print("4.Scrape ImageLink")
    print("5.Exit")
    
    ch=int(input("Enter your choice. =.   "))
    if ch==1:
        paragraph()
        
    elif ch==2:
        headline()
    elif ch==3:
        Weblink()
        
    elif ch==4:
        image()
        
    elif ch==5:
        exit()
        
    else:
        print("Wrong choice")
        
        exit()
```
